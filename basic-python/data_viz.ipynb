## Data Visualization
Very useful data visualization resource: https://python-graph-gallery.com/

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as GridSpec
import numpy as np
```

### Basic plots 

```python
# time points
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
# for more style options - https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
plt.style.use('ggplot')
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

Create multiple plots with `plt.subplots`

```python
# Some example data to display
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Axis [0, 0]')
axs[0, 1].plot(x, y, 'tab:orange')
axs[0, 1].set_title('Axis [0, 1]')
axs[1, 0].plot(x, -y, 'tab:green')
axs[1, 0].set_title('Axis [1, 0]')
axs[1, 1].plot(x, -y, 'tab:red')
axs[1, 1].set_title('Axis [1, 1]')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

### Custom plot layout using `GridSpec`
First initiate figure and gridspec objects separately. Then, use ` add_subplot()` method to create axes objects. Once the axes are generated, use methods like `scatter`, `plot`, etc to visualize data.

```python
# here, figure and gridspec instances are initiated separately
fig = plt.figure(constrained_layout=True)
spec = GridSpec.GridSpec(ncols=2, nrows=2, figure=fig)
f_ax1 = fig.add_subplot(spec[0, 0])
# scatter plot with random x and y
f_ax1.scatter(np.random.rand(100)**2, np.random.rand(100), s= 3)
f_ax2 = fig.add_subplot(spec[0, 1])
f_ax2.scatter(np.random.rand(100)**2, np.random.normal(size=100), s= 3)
f_ax3 = fig.add_subplot(spec[1, 0])
f_ax3.text(0.4, 0.4, s = "Empty plot")
f_ax4 = fig.add_subplot(spec[1, 1])
f_ax4.plot(np.arange(100), np.random.normal(size=100))
```

Use the method `add_gridspec()` for more complex layouts where you can control the number of rows and columns of a grid each sub-figure occupies.

```python
fig3 = plt.figure(constrained_layout=True)
gs = fig3.add_gridspec(3, 3)
f3_ax1 = fig3.add_subplot(gs[0, 1:2])
f3_ax1.set_title('gs[0, 1:2]')
f3_ax2 = fig3.add_subplot(gs[1, :])
f3_ax2.set_title('gs[1, :]')
f3_ax3 = fig3.add_subplot(gs[2, :])
f3_ax3.set_title('gs[2, :]')
```
