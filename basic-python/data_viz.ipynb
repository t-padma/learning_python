## Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

### `matplotlib.pyplot` for data visualization

`matplotlib.pyplot` is a state-based interface to matplotlib. It provides an implicit, MATLAB-like, way of plotting. It also opens figures on your screen, and acts as the figure GUI manager. It is mainly intended for interactive plots and simple cases of programmatic plot generation:

```python
## most basic
x = np.arange(0, 5, 0.1)
y = np.sin(x)
plt.plot(x, y)
```

`plt.subplots()` in Python is similar to `par(mfrow = )` in R.

If we use `plt.subplots(nrow = M, ncol = N, ....)`, it will create a grid of $M \times N$ for $MN$ total plots.

See [subplots documentation.](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html#matplotlib.pyplot.subplots)

`plt.subplots()` returns two objects - `fig` and `ax`

```python
x = np.arange(0, 5, 0.1)
y = np.sin(x)

# if nothing is mentioned, default values of nrow = ncol = 1 is used
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
# we get our 0-th layer i.e. an empty plot with 4 sub-plots
# in this case, axs object is a 2*2 array
fig, axs = plt.subplots(2, 2)
```

```python
print(axs.shape)
```

Common practice to name the returned objects as `fig, ax` or `fig, axs`

```python
# using tuple unpacking for multiple Axes

# here we have two subplots with axis objects ax1 and ax2
fig_a, (ax1, ax2) = plt.subplots(1, 2)

# instead of an array axes object, we have one axis for each subplot
fig_b, ((ax11, ax12, ax13), (ax14, ax15, ax16)) = plt.subplots(2, 3)
```

```python
# simplest plot using default blue line ("b-")
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])

# using red dots and controlling limits of x,y axis
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis((0, 6, 0, 20))
plt.show()
```

```python
# time points
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

Create multiple plots with `plt.subplots`

```python
# Some example data to display
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

# OPTION 1 to visalize data
fig, axs = plt.subplots(2)
fig.suptitle('Vertically stacked subplots')
axs[0].plot(x, y)
axs[1].plot(x, -y)
```

```python
# OPTION 2 to visualize same data
fig, (ax1, ax2) = plt.subplots(2)
fig.suptitle('Vertically stacked subplots')
ax1.plot(x, y)
ax2.plot(x, -y)
```

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Axis [0, 0]')
axs[0, 1].plot(x, y, 'tab:orange')
axs[0, 1].set_title('Axis [0, 1]')
axs[1, 0].plot(x, -y, 'tab:green')
axs[1, 0].set_title('Axis [1, 0]')
axs[1, 1].plot(x, -y, 'tab:red')
axs[1, 1].set_title('Axis [1, 1]')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```
